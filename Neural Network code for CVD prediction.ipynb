```python
# Defining a function to find the statistical relationship with all the categorical variables
def FunctionAnova(inpData, TargetVariable, ContinuousPredictorList):
    from scipy.stats import f_oneway

    # Creating an empty list of final selected predictors
    SelectedPredictors=[]
    
    print('##### ANOVA Results ##### \n')
    for predictor in ContinuousPredictorList:
        CategoryGroupLists=inpData.groupby(TargetVariable)[predictor].apply(list)
        AnovaResults = f_oneway(*CategoryGroupLists)
        
        # If the ANOVA P-Value is <0.05, that means we reject H0
        if (AnovaResults[1] < 0.05):
            print(predictor, 'is correlated with', TargetVariable, '| P-Value:', AnovaResults[1])
            SelectedPredictors.append(predictor)
        else:
            print(predictor, 'is NOT correlated with', TargetVariable, '| P-Value:', AnovaResults[1])
    
    return(SelectedPredictors)
```

```python
# Writing a function to find the correlation of all categorical variables with the Target variable
def FunctionChisq(inpData, TargetVariable, CategoricalVariablesList):
    from scipy.stats import chi2_contingency
    
    # Creating an empty list of final selected predictors
    SelectedPredictors=[]

    for predictor in CategoricalVariablesList:
        CrossTabResult=pd.crosstab(index=inpData[TargetVariable], columns=inpData[predictor])
        ChiSqResult = chi2_contingency(CrossTabResult)
        
        # If the ChiSq P-Value is <0.05, that means we reject H0
        if (ChiSqResult[1] < 0.05):
            print(predictor, 'is correlated with', TargetVariable, '| P-Value:', ChiSqResult[1])
            SelectedPredictors.append(predictor)
        else:
            print(predictor, 'is NOT correlated with', TargetVariable, '| P-Value:', ChiSqResult[1])        
            
    return(SelectedPredictors)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score
import pandas as pd

# Assuming X and y are pandas DataFrame and Series respectively
kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

train_f1_scores = []
val_f1_scores = []

for train_index, val_index in kf.split(X, y):
    # Use .iloc to handle pandas DataFrame indexing
    X_train_fold, X_val_fold = X.iloc[train_index], X.iloc[val_index]
    y_train_fold, y_val_fold = y.iloc[train_index], y.iloc[val_index]

    # Create and train the model on the training fold
    clf = LogisticRegression(C=1, penalty='l2', solver='newton-cg', random_state=42)
    clf.fit(X_train_fold, y_train_fold)

    # Predict on the training and validation fold
    train_predictions = clf.predict(X_train_fold)
    val_predictions = clf.predict(X_val_fold)

    # Calculate F1 scores and store them
    train_f1 = f1_score(y_train_fold, train_predictions, average='weighted')
    val_f1 = f1_score(y_val_fold, val_predictions, average='weighted')

    train_f1_scores.append(train_f1)
    val_f1_scores.append(val_f1)

# Plotting the results
plt.figure(figsize=(12, 6))

# Plot training and validation F1 scores
plt.plot(range(1, 11), train_f1_scores, label='Training F1 Score')
plt.plot(range(1, 11), val_f1_scores, label='Validation F1 Score')

# Adding titles and labels
plt.title('Training and Validation F1 Scores per fold')
plt.xlabel('Fold')
plt.ylabel('Weighted F1 Score')
plt.xticks(np.arange(1, 11, step=1))
plt.legend()

# Show the plot
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score
from xgboost import XGBClassifier

# Initialize models
logreg = LogisticRegression(max_iter=1000, random_state=42)
dtree = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(random_state=42)
gboost = GradientBoostingClassifier(random_state=42)
svm = SVC(random_state=42, probability=True)  # Enable probability for AUROC calculation
knn = KNeighborsClassifier()
nb = GaussianNB()
xgboost = XGBClassifier(random_state=42)
adaboost = AdaBoostClassifier(random_state=42)

# List of models
models = [logreg, dtree, rf, gboost, svm, knn, nb, xgboost, adaboost]

# Dictionary to hold model names and their scores
model_scores = {}

# Train and evaluate each model
for model in models:
    # Train model
    model.fit(X_train, y_train)

    # Predict on test data
    y_pred = model.predict(X_test)

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    # Calculate AUROC
    # For models that use decision_function instead of predict_proba
    if hasattr(model, "decision_function"):
        y_scores = model.decision_function(X_test)
    else:
        y_scores = model.predict_proba(X_test)[:, 1]  # Probability estimates for the positive class
    auc = roc_auc_score(y_test, y_scores)

    # Store the scores in the dictionary
    model_scores[model.__class__.__name__] = {
        'accuracy': accuracy,
        'confusion_matrix': cm,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'auroc': auc
    }

model_scores
```
